```python
# main.py

import torch
from torch.utils.data import DataLoader
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import KFold, StratifiedKFold
from sklearn.model_selection import train_test_split
from data.datasets import CustomDataset, TestDataset, only_2_classes, Dataset2classes
from data.transforms import data_augmentation_transforms, get_biased_transforms, test_transforms_Alexnet, test_transforms_Resnet, train_transforms_Resnet
from models.resnet.resnet import ResNet18, ResNet34, ResNet50, ResNet50pretrained, ResNet101, ResNet152, ResNet34Biased
from models.alexnet.alexnet import Alexnet
from utils.visualization import show_image, show_augmented_images
from train import train, trainKF
import tqdm
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Check for CUDA availability
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")


# Paths
csv_path = 'data/train/train.csv'
image_dir = 'data/train/train'
test_dir = 'data/test/test'



#only_2_classes(csv_path)

torch.cuda.empty_cache()
batch_size = 32
num_folds = 100
num_epochs = 30

kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)


dataset = CustomDataset(csv_path, image_dir, transform=train_transforms_Resnet)

train_dataset, validation_dataset = train_test_split(dataset, test_size=0.1, random_state=41)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
validation_loader = DataLoader(validation_dataset, batch_size=batch_size, shuffle=False)





Model_List = [Alexnet, ResNet18, ResNet34, ResNet50, ResNet101]
entrainement_modeles = True

results = {
    'Alexnet': {'train_losses': [], 'val_losses': [], 'train_accuracies': [], 'val_accuracies': []},
    'ResNet18': {'train_losses': [], 'val_losses': [], 'train_accuracies': [], 'val_accuracies': []},
    'ResNet34': {'train_losses': [], 'val_losses': [], 'train_accuracies': [], 'val_accuracies': []},
    'ResNet50': {'train_losses': [], 'val_losses': [], 'train_accuracies': [], 'val_accuracies': []},
    'ResNet101': {'train_losses': [], 'val_losses': [], 'train_accuracies': [], 'val_accuracies': []}
}

if entrainement_modeles :

    for model_name in Model_List:
        print(f"Training {model_name.__name__}")
        model = model_name().to(device)
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.SGD(model.parameters(), lr=0.01, momentum=0.9, weight_decay=1e-5)
        scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5, verbose=True)
        train_losses, val_losses, train_accuracies, val_accuracies = train(
            num_epochs=num_epochs,
            train_loader=train_loader,
            test_loader=validation_loader,
            model=model,
            criterion=criterion,
            optimizer=optimizer,
            scheduler=scheduler,
            device=device
        )
        # Enregistrement des résultats dans le dictionnaire
        results[model_name.__name__]['train_losses'] = train_losses
        results[model_name.__name__]['val_losses'] = val_losses
        results[model_name.__name__]['train_accuracies'] = train_accuracies
        results[model_name.__name__]['val_accuracies'] = val_accuracies
        torch.cuda.empty_cache()



```

```python
epochs = range(1, num_epochs + 1)

# Couleurs pour chaque modèle
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']

# Créer les graphiques
fig, axs = plt.subplots(2, 5, figsize=(20, 10))  # 2 lignes pour les courbes, 6 colonnes pour chaque modèle

# Tracer les courbes pour chaque modèle
for idx, (model_name, metrics) in enumerate(results.items()):
    # Tracer les pertes
    axs[0, idx].plot(epochs, metrics['train_losses'], label='Train Loss', color=colors[idx], linestyle='-')
    axs[0, idx].plot(epochs, metrics['val_losses'], label='Validation Loss', color=colors[idx], linestyle='--')
    axs[0, idx].set_title(f'{model_name} Losses')
    axs[0, idx].set_xlabel('Epochs')
    axs[0, idx].set_ylabel('Loss')
    axs[0, idx].legend()
    # Tracer les précisions
    axs[1, idx].plot(epochs, metrics['train_accuracies'], label='Train Accuracy', color=colors[idx], linestyle='-')
    axs[1, idx].plot(epochs, metrics['val_accuracies'], label='Validation Accuracy', color=colors[idx], linestyle='--')
    axs[1, idx].set_title(f'{model_name} Accuracies')
    axs[1, idx].set_xlabel('Epochs')
    axs[1, idx].set_ylabel('Accuracy (%)')
    axs[1, idx].legend()

# Ajuster la disposition des sous-graphiques
plt.tight_layout()
plt.show()
```

```python
# Définir les paramètres      non biaisé = 0.5, 0.5
p0 = 0.2
p1 = 0.8

# Obtenir les transformations biaisées


##We create a dataset with a Bias and only 2 classes
dataset_2_classes = Dataset2classes(csv_path, image_dir, transform=test_transforms_Resnet, p0 = 0.5, p1 = 0.5) 
dataset_2_classes, validation_dataset_2_classes = train_test_split(dataset_2_classes, test_size=0.1)
train_loader_2_classes = DataLoader(dataset_2_classes, batch_size=batch_size, shuffle=False)
validation_loader_2_classes = DataLoader(validation_dataset_2_classes, batch_size=batch_size, shuffle=False)   

dataset_2_classes_biased = Dataset2classes(csv_path, image_dir, transform=test_transforms_Resnet, p0 = p0,  p1 = p1)
train_loader_2_classes_biased = DataLoader(dataset_2_classes_biased, batch_size=batch_size, shuffle=False)



model1 = ResNet34Biased().to(device)
model2 = ResNet34Biased().to(device)


optimizer1 = optim.Adam(model1.parameters(), lr=0.01)
optimizer2 = optim.Adam(model2.parameters(), lr=0.01)
scheduler = None
criterion = nn.CrossEntropyLoss()

train_losses1, val_losses1, train_accuracies1, val_accuracies1 = train(num_epochs=num_epochs,train_loader=train_loader_2_classes,test_loader=validation_loader_2_classes,model=model1,criterion=criterion,optimizer=optimizer1,scheduler=scheduler,device=device)
train_losses2, val_losses2, train_accuracies2, val_accuracies2 = train(num_epochs=num_epochs,train_loader=train_loader_2_classes_biased,test_loader=validation_loader_2_classes,model=model2,criterion=criterion,optimizer=optimizer2,scheduler=scheduler,device=device)
```

```python
numerator1_s0 = 0
numerator1_s1 = 0
numerator2_s0 = 0
numerator2_s1 = 0

denominator1_s0 = 0
denominator1_s1 = 0
denominator2_s0 = 0
denominator2_s1 = 0

model1.eval()
model2.eval()


# Loop over the validation dataset
for inputs, labels in validation_loader_2_classes:
    inputs = inputs.to(device)
    labels = labels.to(device)
    
    # Evaluate model outputs
    outputs1 = torch.softmax(model1(inputs), dim=1)
    outputs2 = torch.softmax(model2(inputs), dim=1)
    
    # Calculate P(y_hat = 1 | S = 0) and P(y_hat = 1 | S = 1) for model1
    p_hat1_s0 = torch.mean(outputs1[labels == 0, 1]).item()
    p_hat1_s1 = torch.mean(outputs1[labels == 1, 1]).item()
    
    numerator1_s0 += p_hat1_s0
    numerator1_s1 += p_hat1_s1
    
    denominator1_s0 += 1
    denominator1_s1 += 1
    
    # Calculate P(y_hat = 1 | S = 0) and P(y_hat = 1 | S = 1) for model2
    p_hat2_s0 = torch.mean(outputs2[labels == 0, 1]).item()
    p_hat2_s1 = torch.mean(outputs2[labels == 1, 1]).item()
    
    numerator2_s0 += p_hat2_s0
    numerator2_s1 += p_hat2_s1
    
    denominator2_s0 += 1
    denominator2_s1 += 1

# Compute DI for model1
DI1 = (numerator1_s0 / denominator1_s0) / (numerator1_s1 / denominator1_s1)

# Compute DI for model2
DI2 = (numerator2_s0 / denominator2_s0) / (numerator2_s1 / denominator2_s1)

print(f"DI1: {DI1}")
print(f"DI2: {DI2}")
```

```python
#Test des modèles 
torch.save(model.state_dict(), 'lastmodel.pth')
model.eval()
total_val_loss = 0
correct_val = 0
total_val = 0

with torch.no_grad():
    ProgressBar = tqdm.tqdm(validation_loader, total=len(validation_loader))
    for i, (images, labels) in enumerate(ProgressBar):
        if images is None or labels is None:
            continue

        images = images.to(device)
        labels = labels.to(device)

        outputs = model(images)
        loss = criterion(outputs, labels)
        total_val_loss += loss.item()

        _, predicted = torch.max(outputs.data, 1)
        total_val += labels.size(0)
        correct_val += (predicted == labels).sum().item()

    print(f"Validation loss: {total_val_loss / len(validation_loader)}")
    print(f"Validation accuracy: {100 * correct_val / total_val}")

```

```python
#Création d'un fichier csv pour la soumission Kaggle
model.eval()
predictions = []

test_dataset = TestDataset(test_dir, transform=test_transforms_Resnet)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

with torch.no_grad():
    ProgressBar = tqdm.tqdm(test_loader, total=len(test_loader))
    for i, (images, ids) in enumerate(ProgressBar):

        images = images.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        predicted += 1
        predictions.extend(zip(ids, predicted.cpu().numpy()))

def save_predictions(predictions, output_csv):
    df = pd.DataFrame(predictions, columns=['id', 'label'])
    df.to_csv(output_csv, index=False)

save_predictions(predictions, 'v2.csv')
```
